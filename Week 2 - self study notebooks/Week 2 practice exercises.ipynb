```python
import pandas as pd

gapminder_df = pd.read_csv("gapminder.csv")
display(gapminder_df.head())
```

```python
print(gapminder_df["year"].unique())
```

```python
gapminder_2007_df = gapminder_df.query("year==2007")
display(gapminder_2007_df.head())
```

```python
import seaborn as sns

g = sns.scatterplot(gapminder_2007_df, x="gdpPercap", y="lifeExp", hue="continent")
g.set_xlabel("GDP per Capita")
g.set_ylabel("Life Expectency")
```
